#### Null Values

```python
null_values=["na","n/a","none","null","","nan","nat"]
```

#### Column Values

#### Missing

```python
# Triming
for col in df.columns:
    clean = df[col].astype(str).str.strip()
    if pd.api.types.is_numeric_dtype(df[col]):
        df[col] = pd.to_numeric(clean, errors='coerce')
    elif pd.api.types.is_datetime64_any_dtype(df[col]):
        df[col] = pd.to_datetime(clean, errors='coerce')
    elif pd.api.types.is_bool_dtype(df[col]):
        df[col] = clean.map({'True': True, 'False': False})
    elif pd.api.types.is_categorical_dtype(df[col]):
        df[col] = clean.astype('category')
    else:
        df[col] = clean

# Null Convert
df = df.replace(to_replace=null_values, value=np.nan)

# Missing Values
for col in df.columns:
    if df[col].isnull().any():
        df[col + '_isnull'] = ((~df[col].isnull()) & (df[col] != "Unknown")).astype(int)

# Datetime
dates = [i for i in df.columns if any(j in i for j in ["date","dt","timestamp"])]
times = [i for i in df.columns if any(j in i for j in ["time"])]
for col in dates:
    df[col] = pd.to_datetime(df[col], errors='coerce')
for col in times:
    temp = pd.to_datetime(df[col], format='%H:%M:%S', errors='coerce')
    df[col] = pd.to_datetime('1970-01-01') + (temp - temp.dt.normalize())

# Datetime_isnull 
for col in df.columns:
    if 'isnull' in col.lower() and pd.api.types.is_datetime64_any_dtype(df[col]):
        df[col] = df[col].apply(lambda x: 0 if pd.notna(x) and x == pd.Timestamp('1970-01-01 00:00:00') else 1)
        
# import os
# os.makedirs("D:/ZZ/Clean", exist_ok=True)
```
